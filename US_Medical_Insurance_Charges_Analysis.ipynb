# Converting a Str() to Int() value keeping str() as default value for dict key.

```python
import csv
with open('insurance.csv') as insurance_file:
    str_to_int = csv.DictReader(insurance_file)
    converted=[]
    for i in str_to_int:
        for key,value in i.items():
            if i['sex'] == 'female':
                i['sex'] = 0
                
            elif i['sex'] == 'male':
                i['sex'] = 1
                
            elif i['smoker'] == 'yes':
                i['smoker'] = 1
                
            elif i['smoker'] == 'no':
                i['smoker'] = 0  
        converted.append(i)
fields=['age','sex','bmi','children','smoker','region','charges']
with open('converted_insurance.csv', 'w') as converted_insurance_csv:
    converted_writer = csv.DictWriter(converted_insurance_csv, fieldnames=fields)
    converted_writer.writeheader()   #.writeheader() will write the fieldnames as a first row in our csv file.
    for line in converted:
        converted_writer.writerow(line) 
```

# Insurance_charge for person who smokes and person who dose'nt smoke with their Respective Regions

```python
with open('converted_insurance.csv') as converted_file:
    chevy  = csv.DictReader(converted_file)
    for g in chevy:
        def no_smoking(age,sex,bmi,children,smoker):
                charge = 250 * int(age) - 128 * int(sex) + 370 * float(bmi) + 425 * int(children) + 24000 * int(smoker) - 12500
                if g['smoker'] == '0':
                    print("The Insurance charge For Person Who don't Smoke is: " +str(charge)+ ", Who is Living in the region "+g['region'])
                else: 
                    print("The Insurance charge For Person Who Smoke is: " +str(charge)+ ", Who is Living in the region " +g['region'])
                return charge
        no_smoking(g['age'],g['sex'],g['bmi'],g['children'],g['smoker'])
```
